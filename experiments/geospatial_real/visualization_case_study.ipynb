```python
import numpy as np
from tifffile import imread
import matplotlib.pyplot as plt
import dill as pkl
```

```python
x, y, window_size = 1000, 1000, 50
data_path = '../../data/geospatial_data/durham/durham_ndvi.tif'
img = imread(data_path)
fig, ax = plt.subplots(1, 1, figsize=(5,4))
ax.imshow(img[x-window_size:x+window_size,y-window_size:y+window_size])
```

```python
with open('./results_geospatial_real.pkl','rb') as fp:
    res = pkl.load(fp)

model_names = [
    'linear_with_U', 
    'f_cnn_g_mlp_with_U',
    'f_cnn_g_dk_mlp_with_U',
]
colors = ['blue', 'red', 'orange']
ls = ['--', '-', '-']

fig, ax = plt.subplots(1, 1, figsize=(8,7))
x_coord = np.linspace(0, 50, 51)
for i in range(len(model_names)):
    ax.plot(x_coord, res[model_names[i]][50,50:], color=colors[i], ls=ls[i], label=model_names[i])
ax.legend()
```
